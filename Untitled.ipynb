```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("this is x")
plt.ylabel("this is y")
plt.title("this is scatter")
```

# Ans4

```python
# A bar plot or bar chart is a graph that represents the category of data 
# with rectangular bars with lengths and heights that is proportional to the values which they represent. 
# The bar plots can be plotted horizontally or vertically.
```

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.show()
```

```python
# A Box Plot is also known as Whisker plot is created to display the 
# summary of the set of data values having properties like minimum, 
# first quartile, median, third quartile and maximum. In the box plot
# , a box is created from the first quartile to the third quartile, a vertical
# \line is also there which goes through the box at the median.
```

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
plt.box(box1,box2)
plt.show()
```
